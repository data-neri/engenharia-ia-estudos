```python
#tentando_criar_local
#python teste.py
from openai import OpenAI 
   
client = OpenAI(
  base_url="http://127.0.0.1:1234/v1",
  api_key="1234"
)

numero_dias = 7
numero_crianças=2
atividade="música"

prompt=f"Crie um roteiro de viagem de {numero_dias} dias, para uma familia com {numero_crianças} crianças presentes nessa viagem, que gosta de{atividade} "

resposta = client.chat.completions.create(
    model="mistralai/mistral-nemo-instruct-2407",
    messages=[
        {
            "role": "system",
            "content": "Você é um assistende de roteiro de viagem"
        },
        {
            "role": "user",
            "content": prompt,
        }
    ],
    temperature=1.0,
    max_tokens=1000,
    stream=True,
    reasoning_effort="medium",
    
)
# Em vez de print(resposta.choices[0].message.content)
for chunk in resposta:
    if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end="", flush=True)
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os 

# Carrega as variáveis do.env corrigido
load_dotenv()

# Pega os valores do.env
api_base = os.getenv("OPENAI_API_BASE")
api_key = os.getenv("OPENAI_API_KEY")

# Configura o modelo corretamente para o LM Studio
# Não use 'stream=True' aqui dentro
modelo = ChatOpenAI(
    model="local-model",         # Nome do modelo carregado no LM Studio 
    base_url=api_base,           # Usa o endereço do seu.env 
    api_key=api_key,             # Usa a chave 'lm-studio' do seu.env 
    temperature=0.7,
    max_completion_tokens=1000
)

numero_dias = 7
numero_crianças = 2
atividade = "música"

prompt = f"Crie um roteiro de viagem de {numero_dias} dias para uma família com {numero_crianças} crianças que gosta de {atividade}."

# Para o efeito de "escrever pouco a pouco", use o método.stream()
# Isso resolve o aviso (Warning) que apareceu no seu console 
for chunk in modelo.stream(prompt):
    print(chunk.content, end="", flush=True)
```
